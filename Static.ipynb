```python
import pandas as pd
import altair as alt
import numpy as np
import pprint
```

```python
df=pd.read_csv("/Users/yuanlizhu/Desktop/w23/548/individual project/bias_data.csv")
df.sample(5)
```

```python
alt.Chart(df).mark_bar().encode(
    x=alt.X("Disability:N", axis=alt.Axis(title=None, labels=False, ticks=False)),
    y=alt.Y("Rate:Q", axis=alt.Axis(grid=False)),
    color="Disability:N",
    opacity="Age",
    column=alt.Column("Condition:N", header=alt.Header(title=None, labelOrient='bottom'))
).configure_view(
    stroke='transparent'
).configure_header(labelOrient='bottom',
                   labelPadding = 3).configure_facet(spacing=50
)
```

```python
base = alt.Chart(df).mark_bar().encode(
    y=alt.Y("Disability:N", axis=alt.Axis(title=None, labels=False, ticks=False)),
    x=alt.X("Rate:Q", axis=alt.Axis(), scale=alt.Scale(domain=[0, 100])),
    color="Disability:N",
    row=alt.Row("Condition:N", header=alt.Header(title=None, labelOrient='left'))
)


g1 = base.transform_filter(
    alt.datum.Age=='18~44'
).properties(width=250,height=70,title="Age Group: 18~44")

g2 = base.transform_filter(
    alt.datum.Age=='45~64'
).properties(width=250,height=70,title="Age Group: 45~64")


g3 = base.transform_filter(
    alt.datum.Age=='64~'
).properties(width=250,height=70,title="Age Group: 65~")


(g1 | g2 | g3).resolve_scale(
    y='shared'
).properties(
    title='General Health and Chronic Conditions of People with and without Disabilities' # set the title of the chart
).configure_title(
    fontSize=14, # set the font size of the title
    anchor='middle' # set the anchor of the title to the middle
).configure_legend(
    labelFontSize=12  # set the font size of the legend labels
)
```
